```python
from skimage import morphology, filters
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['image.cmap'] = 'gray'
```

```python
# Creating image
dimensions=(378,378)
im = np.random.normal(size=(dimensions))
```

```python
# Smoothing
diskSize = 2
disk = morphology.disk(diskSize)
imSmooth = filters.median(im, disk)
```

```python
# Thresholding
threshold = 0.3
# threshold = filters.threshold_otsu(imSmooth)
imThresh = imSmooth > threshold
```

```python
# Smoothed image
fig, ax = plt.subplots(1,2)
ax[0].imshow(imSmooth)
ax[0].set_title('Smoothed image')
ax[0].axis('off')
ax[1].imshow(imThresh)
ax[1].set_title('Thresholded image')
ax[1].axis('off')

plt.tight_layout()
plt.show()
```
