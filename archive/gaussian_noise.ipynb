```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from src.preprocessing import process_image, process_video, preprocess, resize
from src.distance_measures import analyse_video
from src.visualizer import visualize
from src.gaussian_noise import generate_noise
from skimage.color import label2rgb
plt.rcParams['image.cmap'] = 'gray'
```

```python
# Creating image
im_size = 1002
threshold = 0.2
smooth = 2
size = 400

ims = generate_noise(size=im_size, threshold=threshold, smooth=smooth)
im = ims[0]
im = resize(arr=im, size=size)
```

```python
plt.imshow(im != 0)

plt.tight_layout()
plt.show()
```

```python
# Smoothed image
fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].imshow(im)
ax[0].set_title('Smoothed image')
ax[0].axis('off')
ax[1].imshow(im != 0)
ax[1].set_title('Thresholded image')
ax[1].axis('off')
ax[2].imshow(label2rgb(im, bg_label=0))
ax[2].set_title('Labeled image')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
